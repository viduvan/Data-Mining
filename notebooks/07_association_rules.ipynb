# Khai phá Dữ liệu Tuyển sinh Đại học Việt Nam (Data Mining)
## Notebook 07: Khai phá Luật kết hợp (Association Rule Mining - Apriori)

Notebook này thực hiện:
1. Discretize (rời rạc hóa) các trường dữ liệu liên tục thành các danh mục.
2. Biến đổi dữ liệu thành các giao dịch (Transactions).
3. Chạy thuật toán Apriori để tìm các tập phổ biến (Frequent Itemsets).
4. Trích xuất các luật kết hợp có ý nghĩa thống kê dựa trên Support, Confidence, và Lift.

```python
import pandas as pd
import numpy as np
from src.mining.association_rules import AssociationRuleMiner

df = pd.read_csv("../data/processed/admission_processed.csv", encoding="utf-8-sig")
```

### 1. Chuẩn bị dữ liệu và khởi tạo Miner

```python
miner = AssociationRuleMiner(df)
df_transactions = miner.prepare_transaction_data()
print(f"Kích thước dữ liệu giao dịch: {df_transactions.shape}")
df_transactions.head()
```

### 2. Khai phá luật kết hợp bằng Apriori

```python
rules = miner.mine_rules(min_support=0.05, min_confidence=0.5)
print(f"Tìm thấy {len(rules)} luật kết hợp thỏa mãn điều kiện.")
```

### 3. Top 20 luật kết hợp có độ tin cậy cao nhất

```python
miner.print_rules(rules, top_n=20)
```
