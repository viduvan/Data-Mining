## Notebook 06: Phân cụm trường Đại học (K-Means Clustering)

Notebook này thực hiện:
1. Phân nhóm các trường Đại học dựa trên các đặc điểm tuyển sinh chính.
2. Xác định số lượng cụm tối ưu K bằng phương pháp Elbow và Silhouette.
3. Giảm chiều dữ liệu bằng PCA để trực quan hóa cụm trường.
4. Phân tích đặc điểm và định danh hồ sơ từng cụm.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from src.mining.clustering import ClusteringAnalyzer

plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_theme(style="whitegrid")

df = pd.read_csv("../data/processed/admission_processed.csv", encoding="utf-8-sig")
```

### 1. Khởi tạo bộ phân cụm và tìm số cụm K tối ưu

```python
analyzer = ClusteringAnalyzer(df)
# Vẽ biểu đồ Elbow
analyzer.plot_elbow(max_k=8)
```

### 2. Tiến hành phân cụm với K=4

```python
df_schools = analyzer.cluster_schools(k=4)
print(df_schools[['school_name', 'cluster_label', 'avg_score', 'num_majors', 'total_quota']].head(15))
```

### 3. Trực quan hóa các cụm trường trên không gian 2D PCA

```python
analyzer.plot_clusters_2d()
```

### 4. Phân tích đặc trưng hồ sơ (Centroid Profiles) của các cụm

```python
# Tính trung bình các thuộc tính theo từng cụm
features = ['avg_score', 'max_score', 'min_score', 'num_majors', 'total_quota']
cluster_profiles = df_schools.groupby('cluster_label')[features].mean().reset_index()
print("Hồ sơ đặc điểm từng cụm trường (Centroids):")
print(cluster_profiles.round(2))
```
