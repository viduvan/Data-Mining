Notebook này thực hiện:
1. Nhập điểm thi của thí sinh cùng tổ hợp xét tuyển.
2. Hệ thống tính tổng điểm xét tuyển đã cộng ưu tiên khu vực.
3. Lọc và tính điểm Safety Score cho tất cả các ngành học thỏa mãn.
4. Phân loại kết quả gợi ý theo 4 mức an toàn và sắp xếp bảng xếp hạng.

```python
import pandas as pd
import numpy as np
from src.recommendation.recommender import AdmissionRecommender

# Khởi tạo Recommender từ file CSV đã processed
recommender = AdmissionRecommender.from_csv("../data/processed/admission_processed.csv")
```

### 1. Thử nghiệm gợi ý cho học sinh thi khối A00 đạt 24.5 điểm (Toán 8.0, Lý 8.5, Hóa 8.0)

```python
scores = [8.0, 8.5, 8.0]
subject_group = "A00"
priority_region = "KV2"

results = recommender.recommend(
    scores=scores,
    subject_group=subject_group,
    priority_region=priority_region,
    top_n=15
)
recommender.print_results(results)
```

### 2. Thử nghiệm gợi ý cho học sinh thi khối D01 đạt 26 điểm (Toán 9.0, Văn 8.5, Anh 8.5) và lọc theo nhóm ngành Kinh tế

```python
scores = [9.0, 8.5, 8.5]
subject_group = "D01"
priority_region = "KV3"
major_group = "Kinh tế - Quản trị"

results_kt = recommender.recommend(
    scores=scores,
    subject_group=subject_group,
    priority_region=priority_region,
    major_group=major_group,
    top_n=10
)
recommender.print_results(results_kt)
```
